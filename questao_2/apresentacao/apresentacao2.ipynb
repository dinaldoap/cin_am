## Comparação dos classificadores
### Teste de Friedman
<ul>
    <li>Hipóteses:
        <ul>
            <li>Hipótese nula: as diferenças entre os classificadores para o conjunto de dados utilizado são insignificantes</li>
            <li>Hipótese alternativa: as diferenças entre os classificadores para o conjunto de dados utilizado são estatisticamente significantes.</li>
        </ul>
    </li>    
</ul>

<ul>
    <li>Resultado:
        <ul>
            <li>p-value = 9.4194712315349006e-36</li>
            <li>Rejeição da hipótese nula com p-value &lt; 0.05.</li>
